```python
import json
import matplotlib.pyplot as plt
import matplotlib
#  JSON 
with open('./multi_backend_model_data/HLD.json', 'r') as f:
    data = json.load(f)

#  'lasso_model'  'ElasticNet_model'
data.pop('lasso_model')
data.pop('ElasticNet_model')

# 
plt.figure(figsize=(10, 6))
model_name = ['random_forest','Linear_model','Ridge_model','SVR_model','KNN_model','noisy_error']
key_name = ['QEMOS','Linear','Ridge','SVR','KNN','Unmitigated']
matplotlib.rcParams['font.family'] = 'Times New Roman'
matplotlib.rcParams['font.size'] = 11
data_1 = {}
for index,key in enumerate(model_name):
    data_1[key_name[index]] = data[key]
#  key-value 
colors = ['#FF0000', '#1E90FF','#8A2BE2' , 'grey', '#228B22', '#A0522D']  #  RGB 
for i, (key, value) in enumerate(data_1.items()):
    plt.plot(range(2, len(value) + 2), value, marker='o', label=key, color=colors[i])
plt.axvline(x=10, linestyle='--', color='dodgerblue')
# 
# plt.title('Circuit Error by Qubits')
plt.xlabel('Qubits')
plt.ylabel('Circuit Error')
plt.xticks(range(2, len(value) + 2))  # 
plt.legend(loc='upper left', fontsize='small')  # 
plt.savefig('./1/1.svg', dpi=1000)
# plt.grid(True)  # 
plt.show()  # 
```

```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
#  JSON 
with open('./multi_backend_model_data/HLD.json', 'r') as f:
    data = json.load(f)

#  'lasso_model'  'ElasticNet_model'
model_name = ['tianyan_mited','tianyan_error']
key_name = ['QEMOS','Unmitigated']
data_1 = {}
for index,key in enumerate(model_name):
    data_1[key_name[index]] = data[key]
# 
#  DataFrame
df = pd.DataFrame(data_1)
# Times New Roman
matplotlib.rcParams['font.family'] = 'Times New Roman'
matplotlib.rcParams['font.size'] = 11
#  CSV 
df.to_csv('./1/data_table.csv', index=False)
plt.figure(figsize=(10, 6))

#  key-value 
colors = [ '#FF0000','#1E90FF','#A0522D',  'grey', '#228B22', '#8A2BE2']  #  RGB 
for i, (key, value) in enumerate(data_1.items()):
    if (key!='Qubits'):
        plt.plot(range(2, len(value) + 2), value, marker='o', label=key, color=colors[i])
plt.axvline(x=10, linestyle='--', color='dodgerblue')

# 
# plt.title('Compare the abilities of QRFMqbits and Q_learn on random circuits')
plt.xlabel('Qubits',fontdict={'family': 'Times New Roman', 'size': 12})
plt.ylabel('Circuit Error',fontdict={'family': 'Times New Roman', 'size': 12})
plt.xticks(range(2, len(value) + 2))  # 
# plt.legend(loc='upper left', fontsize='small')  # 
# plt.grid(True)  # 
plt.legend(loc='upper left', prop={'family': 'Times New Roman', 'size': 11})
plt.savefig('./1/1.svg', dpi=1000)
plt.show()  # 
```
