```python
import json
import matplotlib.pyplot as plt
import matplotlib
# 读取 JSON 文件
with open('./multi_backend_model_data/HLD.json', 'r') as f:
    data = json.load(f)

# 删除 'lasso_model' 和 'ElasticNet_model'
data.pop('lasso_model')
data.pop('ElasticNet_model')

# 创建一个新的图形
plt.figure(figsize=(10, 6))
model_name = ['random_forest','Linear_model','Ridge_model','SVR_model','KNN_model','noisy_error']
key_name = ['QEMOS','Linear','Ridge','SVR','KNN','Unmitigated']
matplotlib.rcParams['font.family'] = 'Times New Roman'
matplotlib.rcParams['font.size'] = 11
data_1 = {}
for index,key in enumerate(model_name):
    data_1[key_name[index]] = data[key]
# 对每个 key-value 对进行循环
colors = ['#FF0000', '#1E90FF','#8A2BE2' , 'grey', '#228B22', '#A0522D']  # 使用 RGB 颜色值
for i, (key, value) in enumerate(data_1.items()):
    plt.plot(range(2, len(value) + 2), value, marker='o', label=key, color=colors[i])
plt.axvline(x=10, linestyle='--', color='dodgerblue')
# 添加标题和标签
# plt.title('Circuit Error by Qubits')
plt.xlabel('Qubits')
plt.ylabel('Circuit Error')
plt.xticks(range(2, len(value) + 2))  # 设置横坐标标签
plt.legend(loc='upper left', fontsize='small')  # 添加图例
plt.savefig('./1作图/选择模型1.svg', dpi=1000)
# plt.grid(True)  # 添加网格线
plt.show()  # 显示图形
```
